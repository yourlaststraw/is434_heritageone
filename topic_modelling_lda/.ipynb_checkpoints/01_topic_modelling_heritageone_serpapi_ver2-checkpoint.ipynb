```python
import csv
import requests

# SerpApi configuration (Replace with your SerpApi API key and Place ID)
API_KEY = "f534d8a01a03b6b8a0d8f4b76b60fe8c3b9774a72cf79ff645803ba9d086f44b"
PLACE_ID = "ChIJbwht1dk82jERGOoy1353_1E"
BASE_URL = "https://serpapi.com/search"

# CSV header
header = ['link', 'rating', 'date', 'user_name', 'description', 'likes']

# Function to fetch reviews from SerpApi
def fetch_reviews(page_token=None):
    # Set the parameters for the initial request
    params = {
        'engine': 'google_maps_reviews',
        'place_id': PLACE_ID,
        'hl': 'en',
        'api_key': API_KEY
    }
    
    # If there's a next_page_token, include it in the request and add 'num'
    if page_token:
        params['next_page_token'] = page_token
        params['num'] = '20'  # Add num only when paginating
    
    # Make the request to SerpApi
    response = requests.get(BASE_URL, params=params)
    return response.json()

# Function to extract reviews and write to CSV
def write_reviews_to_csv(reviews, csv_writer):
    for review in reviews:
        csv_writer.writerow([
            review.get('link'),
            review.get('rating'),
            review.get('date'),
            review.get('user', {}).get('name'),
            review.get('snippet'),
            review.get('likes')
        ])

# Main function to paginate through reviews
def scrape_all_reviews():
    next_page_token = None
    first_page = True
    
    with open('google_maps_reviews_all_pages.csv', 'w', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        
        while first_page or next_page_token:
            # Fetch reviews (pass the next_page_token if available)
            data = fetch_reviews(page_token=next_page_token)
            
            # Extract reviews from the response
            reviews = data.get('reviews', [])
            if reviews:
                # Write reviews to CSV
                write_reviews_to_csv(reviews, writer)
                print(f"Fetched {len(reviews)} reviews")
            
            # Get the next_page_token for the next iteration
            next_page_token = data.get('serpapi_pagination', {}).get('next_page_token')
            
            # Set first_page to False after the first iteration
            first_page = False

    print("Scraping completed. All reviews are saved in 'google_maps_reviews_all_pages.csv'.")

# Run the scraper
scrape_all_reviews()
```
